Problem 2
The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information; (Provide an appropriate visual representation of your answers, without which marks will be deducted)

2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import scipy.stats as stats
```

```python
#since the data is normally distributed,importing norm.
from scipy.stats import norm
```

```python
#given mean is 5 and standard deviation is 1.5
mean,sigma = 5,1.5
```

```python
#2.1 The proportion of gunny bags having a breaking strength less than 3.17 kg per sq cm
x=3.17
z1= (x-mean)/sigma
z1=(3.17-5)/1.5
z1
```

```python
p=stats.norm.cdf(z1)
p
```

```python
#Alternate way
#2.1 The proportion of gunny bags having a breaking strength less than 3.17 kg per sq cm
round(stats.norm.cdf(3.17,loc=5,scale=1.5),4)
```

```python
#ploting the normal distribution
plt.figure(figsize=(6,3))
x_axis = np.arange(0,10,0.001)
y_axis = stats.norm.pdf(x_axis,5,1.5)
plt.plot(x_axis,y_axis,label="Normal Distribution")
plt.axvline(3.17,color="r",label = "Breaking Strength=3.17")
plt.fill_between(x_axis,y_axis,where=x_axis<3.17,color="green")
plt.xlabel("Breaking Strength (kg per sq cm)")
plt.ylabel("Probability Density")
plt.title("Breaking Strength<3.17")
plt.legend(loc="upper right",fontsize=8)
plt.show()
```

Hence the prportion of gunny bags having breakingstrength less than 3.17 kg per sq cm is 11.12%

```python
#2.2 The proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.
round(1-stats.norm.cdf(3.5,loc=5,scale=1.5),4)
```

```python
#ploting the normal distribution
plt.figure(figsize=(6,3))
x_axis = np.arange(0,10,0.001)
y_axis = stats.norm.pdf(x_axis,5,1.5)
plt.plot(x_axis,y_axis,label="Normal Distribution")
plt.axvline(3.6,color="r",label = "Breaking Strength=3.6")
plt.fill_between(x_axis,y_axis,where=x_axis>3.6,color="green")
plt.xlabel("Breaking Strength (kg per sq cm)")
plt.ylabel("Probability Density")
plt.title("Breaking Strength of atleast 3.6")
plt.legend(loc="upper right",fontsize=8)
plt.show()
```

Hence the proportion of gunny bags having a breaking strength of atleast 3.6 kg per sq cm is 84.13%

```python
#2.3 The proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.
round(stats.norm.cdf(5.5,loc=5,scale=1.5)-stats.norm.cdf(5,loc=5,scale=1.5),4)
```

Hence the proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm is 13.06%

# 4.1 How does the hardness of implants vary depending on dentists?
 to know how the hardness of implants vary on depending on dentists we have to perform a one way anova test.
 null hypothesis, The mean hardness is same for all the dentists
H0,mu1=mu2=mu3=mu4=mu5
alternate hyothesis, Atleast one of the mean hardness is different
HA,mu1=!=mu2=!mu3=!mu4=!mu5
Assumptions for one way annova test
 It should follow a normal distribution- Can be checked through Shapiro-Wilks test.
 Varainces should be equal - Can be checked through Levenes test.

Shapiro Wilks Test

alternate hypothesis, Ha = hardness does not follow a normal distribution

```python
from scipy.stats import shapiro
p_value =stats.shapiro(df["Response"])
print(p_value)
```

Levenes Test
null hypotheis , H0 = all the population variances are equal
alternate hypothesis, Ha = atleast one of the variance is different from the rest

```python
from scipy.stats import levene
p_value =levene(df['Response'][df['Dentist']==1],
           df['Response'][df['Dentist']==2],
           df['Response'][df['Dentist']==3],
           df['Response'][df['Dentist']==4],
           df['Response'][df['Dentist']==5])
print(p_value)
```

```python
#since the p value is less than 0.05 therefore we reject the null hpothesis and atleast one of the variance is different.
```
